### Create 500 random samples from normal distribution with mean 1 and  standard deviation 0.5

```python
mmean = 1
mstd = 0.5
randon_sample = np.random.normal(mmean,scale=mstd,size=500)
print(randon_sample)
```

###  Plot the histogram of the above normal distribution 

```python
plt.hist(randon_sample)
```

```python
count, bins, ignored = plt.hist(randon_sample, 20, normed=True)

# Plot the distribution curve
plt.plot(bins, 1/(mstd * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mmean)**2 / (2 * mstd**2) ),       linewidth=3, color='y')
plt.show()
```
